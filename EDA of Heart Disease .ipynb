# Check Chest Pain Type

Chest pain type (4 values)
- value 0: typical angina
- value 1: atypical angina
- value 2: non-anginal pain
- value 3: asymptomatic

```python
sns.countplot(y = data['cp'],data = data)
plt.yticks([0,1,2,3],['Typical Angina','Atypical Angina','Non-Anginal','Asymptomatic'])
plt.ylabel('Chest pain')
plt.show() 
```

# Plot Continuous Variables

```python
category_variable = []
continuous_variable = []

for column in data.columns:
    if data[column].nunique() <= 10:
        category_variable.append(column)
    else:
        continuous_variable.append(column)
```

```python
category_variable
```

```python
continuous_variable
```

```python
data.hist(continuous_variable, figsize = (20,20))
plt.tight_layout()
plt.show()
```
